```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.ebay.fr/b/Telephones-mobiles/9355/bn_16576651"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36",
    "Accept-Language": "en-US, en;q=0.5",
}

html = requests.get(url, headers=headers)
soup = BeautifulSoup(html.text, "lxml")

result_list = []

# Find all product items on the page
listings = soup.find_all("div", class_="s-item__info clearfix")
images = soup.find_all("div", class_="s-item__wrapper clearfix")

# Loop through listings and images (assuming they have the same length)
for listing, image_container in zip(listings, images):
    # Check if title element exists before accessing text
    title_element = listing.find("div", class_="s-item__title")
    title = title_element.text if title_element is not None else ""  # Set an empty string if not found

    # Similar check for price element
    price_element = listing.find("span", class_="s-item__price")
    price = price_element.text if price_element is not None else ""

    product_url = listing.find("a")
    link = product_url["href"] if product_url is not None else ""  # Set an empty string if not found

    product_status_element = listing.find("div", class_="s-item__subtitle")
    product_status = (
        product_status_element.text
        if product_status_element is not None
        else "No status available"
    )

    # Only process listings with both title and price
    if title and price:
        title_text = title.strip()
        price_text = price.strip()
        status = product_status.strip()

        # Find image within the image container
        image = image_container.find("img")
        image_url = image["src"] if image is not None else ""  # Set an empty string if not found

        result_dict = {
            "title": title_text,
            "price": price_text,
            "image_url": image_url,
            "status": status,
            "link": link,
        }
        result_list.append(result_dict)

# Print the extracted information
for item in result_list:
    print(f"Title: {item['title']}")
    print(f"Price: {item['price']}")
    print(f"Image URL: {item['image_url']}")
    print(f"Status: {item['status']}")
    print(f"Product Link: {item['link']}")
    print("-------------------------")
```

```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

#  Specify the URL of webpage you want to scrape
url = 'https://www.ebay.fr/b/Telephones-mobiles/9355/bn_16576651'
# Send the HTTP request to the specified URL and save the response
r = requests.get(url)
#  Get the HTML content of webpage
soup = BeautifulSoup(r.content,'html.parser')

# Find the parent element.
info = soup.find_all('div',{'class':'s-item__info clearfix'})

titles=[];links=[];prices = [];products=[];livraison=[]
# Iterate parent element and get required data from that
for item in info:
    title= item.find('h3',{'class':'s-item__title'}).text
    price= item.find('span',{'class':'s-item__price'}).text.replace('EUR','').replace(',','')
    frais_livraison= item.find('span',{'class':'s-item__shipping s-item__logisticsCost'}).text
    link= item.find('a',{'class':'s-item__link'})['href']

    
    titles.append(title)
    prices.append(price)
    links.append(link)
    livraison.append(frais_livraison)
products=pd.DataFrame({
    'titre':titles,
    'prix':prices,
    'frais_livraison':livraison,
    'lien':links
    })
# Save the all scraped data into csv
products.to_csv("Products.csv",index=False)
```
